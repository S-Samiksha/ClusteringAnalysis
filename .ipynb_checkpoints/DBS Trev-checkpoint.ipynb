```python
import numpy as np
import pandas as pd
import seaborn as sb
import matplotlib.pyplot as plt # we only need pyplot
import datetime
sb.set() # set the default Seaborn style for graphics
```

```python
data = pd.read_csv("miami_housing.csv")
data.head()
```

```python
fig = plt.gcf()
fig.set_size_inches(17, 8)

sb.scatterplot(data=data, x='LATITUDE', y='LONGITUDE')
```

```python
data_trim = data[['LATITUDE','LONGITUDE']]
data_trim.head()
```

```python
# finding epsilon and minPts parameters using K-NN
# approx minPts = 4 for 2-D dataset, k = minPts.

from sklearn.neighbors import NearestNeighbors
nbrs = NearestNeighbors(n_neighbors=4).fit(data_trim)

neigh_dist, neigh_ind = nbrs.kneighbors(data_trim)
sort_neigh_dist = np.sort(neigh_dist, axis=0)
```

```python
fig = plt.gcf()
fig.set_size_inches(17, 8)

k_dist = sort_neigh_dist[:, 3]
plt.plot(k_dist)
plt.show()
```

```python
from sklearn.cluster import DBSCAN
from plotnine import *
```

```python
#eps = approx 0.008
clust = DBSCAN(eps=0.008, min_samples=4).fit(data_trim)
labelList = ["Noise"]
labelList = labelList + ["Cluster " + str(i) for i in range(1, len(set(clust.labels_)))]


data_trim["assignments"] = clust.labels_
(ggplot(data_trim, aes(x='LATITUDE', y='LONGITUDE', color='factor(assignments)')) + 
geom_point() + theme_minimal() + scale_color_discrete(name = 'Cluster Assignment',
                                                     labels = labelList) +
theme(panel_grid_major = element_blank()) + 
labs(title = 'DBScan, eps=0.008, min_sam=4'))
```
